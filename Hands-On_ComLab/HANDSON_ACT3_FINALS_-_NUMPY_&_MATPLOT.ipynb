```python
# Matrices

# Matrix 1: Practice per hours
practice_matrix = np.array([
    [8, 8, 6, 9],
    [4, 6, 9, 5],
    [5, 2, 3, 3],
    [6, 8, 4, 6],
    [3, 4, 5, 6]
])

# Matrix 2: Improvement percent
improvement_matrix = np.array([
    [10, 12,  8,  9],
    [14, 18, 20, 15],
    [11,  9, 10, 12],
    [16, 19, 14, 17],
    [9, 11,  12, 13] 
])

# Matrix 3: Skill difficulty
difficulty_matrix = np.array([
    [6, 7, 8, 5],
    [5, 6, 9, 4],
    [7, 6, 5, 6],
    [8, 5, 6, 7],
    [6, 7, 4, 8]  
])

# Matrix 4: Performance score
performance_matrix = np.array([
    [70, 75, 80, 85],
    [78, 82, 88, 90],
    [72, 70, 76, 80],
    [85, 88, 90, 92],
    [80, 83, 85, 87] 
])
```

```python
# Bar Graph

plt.bar(range(len(practice_matrix[3])), practice_matrix[3])
plt.title("Practice Hours (2024)")
plt.xlabel("Categories")
plt.ylabel("Hours")

plt.show()
```

```python
# Line Graph

plt.plot(improvement_matrix[4], marker='o')
plt.title("Improvement Percentage (2025)")
plt.xlabel("Categories")
plt.ylabel("Percentage")
plt.show()
```

```python
# Scatter Plot

plt.scatter(range(len(difficulty_matrix[2])), difficulty_matrix[2])
plt.title("Difficulty Index (2022)")
plt.xlabel("Categories")
plt.ylabel("Difficulty Level")

plt.show()
```

```python
# Heatmap

plt.imshow(practice_matrix, cmap='Blues')
plt.title("Practice Hours by Year")
plt.yticks(range(len(years)), years)  # Add years on the side
plt.colorbar()


plt.show()
```
